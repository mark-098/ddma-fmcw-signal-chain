#FMCW radar data processing scripts
This repository is an FMCW radar pipeline for processing DDMA MIMO ADC data acquiered from the AWR2944EVM with the aid of the DCA1000EVM capture module.

The public demo code is designed to perform processing from raw ADC data, up to RDM domain tracking and automatic classification feature extraction. Under an assumption of a single track per frame and for stationary radars only.

The data used within this demo may be found under: [LINK]

This repository is in development, hence not all docuemnation may reflect the current code status.

## Demo usage

The current demo output control is though basic ```boolean``` parameteres inside the ```ddma_chain_main.py``` file's ```if __name__ == "__main__":``` section.

In addtiion the capture settings and achived velocity and range resolutions must be configured there.

Processing settings, such as:

*   Interference mitigation control,
*   CFAR control,
*   DBSCAN settings,
*   KF tracker,

are controlled from within the ```ddma_dataclasses.py``` file.
